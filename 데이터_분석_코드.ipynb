# 날씨 데이터 추가

### 1. 교통수단으로 인한 사고는 날씨 및 지형과 밀접한 관계가 있다고 판단

### 2. 기상청 기상자료개발개방포털(https://data.kma.go.kr)의 종관기상관측(ASOS) 자료 활용

### 3. 대전광역시, 세종특별자치시, 충청남도의 2020/01/01 ~ 2023/01/31 기간의 일간 '일강수량(mm)', '일 최심적설(cm)', '안개 계속시간간(hr)' 데이터 활용

### 4. 날씨 데이터의 '일시' 와 사건 데이터의 사건 발생 날짜('date')가 일치하고 날씨 데이터의 '지점명' 값이 사건발생 주소('HPPN_PNU_ADDR')에 포함되는 경우 해당 사건 데이터의 행에 '일강수량(mm)', '일 최심적설(cm)', '안개 계속시간(hr)' 데이터 추가

### 5-1. 해당주소의 날씨 데이터가 존재하지 않는 경우 좌표 데이터를 바탕으로 해당 날짜의 가장 가까운 곳의 날씨를 Euclidean Distance 공식을 이용해서 가져온다. 해당 방법은 정확도가 높지만 속도가 느리고 데이터의 크기가 작은 경우 유용하다. 

### 5-2. 해당주소의 날씨 데이터가 존재하지 않는 경우 같은 날짜의 다른 지역의 날씨를 랜덤으로 가져온다. 랜덤으로 가져오는 이유는 순서대로 탐색하면 처음 나오는 값만 계속 불러오게게 된다. 이는 데이터의의 정확도를 감소시키고 모델링 과정에서 OverFitting을 발생시킬 수 있다. 해당 방법은 정확도가 떨어지지만 속도가 빠르고 데이터의 크기가 큰 경우 유용하다.

## - KP2020

```python
kp2020_weather = pd.read_csv('/content/drive/MyDrive/경찰청/2020날씨.csv',encoding='cp949')
```

```python
kp2020_weather['일강수량(mm)'].fillna(0, inplace=True)
kp2020_weather['일 최심적설(cm)'].fillna(0, inplace=True)
kp2020_weather['안개 계속시간(hr)'].fillna(0, inplace=True)
```

```python
#날씨 데이터를 장소와 날짜에 맞게 병합 

import re
from datetime import datetime

def extract_date_from_df(df, column_name):
    return [datetime.strptime(x.split(" ")[0], '%y/%m/%d').date() for x in df[column_name]]
f_df_2020['date'] = extract_date_from_df(f_df_2020, 'RECV_CPLT_DM')

f_df_2020['일강수량(mm)'] = np.nan
f_df_2020['일 최심적설(cm)'] = np.nan
f_df_2020['안개 계속시간(hr)'] = np.nan
f_df_2020.reset_index(drop=True)

for index1, row1 in f_df_2020.iterrows():
    for index2, row2 in kp2020_weather.iterrows():
        if str(row1['date']) == str(row2['일시']) and row2['지점명'] in row1.HPPN_PNU_ADDR:
            f_df_2020.at[index1, '일강수량(mm)'] = row2['일강수량(mm)']
            f_df_2020.at[index1, '일 최심적설(cm)'] = row2['일 최심적설(cm)']
            f_df_2020.at[index1, '안개 계속시간(hr)'] = row2['안개 계속시간(hr)']
```

```python
#날씨 데이터가 장소와 날짜에 맞게 존재하지 않는 경우 가장 가까운 거리를 계산해서 해당 날씨 정보를 가져온다.

def get_nearest_row(df, index, column_names):
    min_distance = float('inf')
    nearest_row = None
    
    for i, row in df[~df.index.isin([index])].iterrows():
        distance = ((row['HPPN_X'] - df.loc[index, 'HPPN_X']) ** 2 + 
                    (row['HPPN_Y'] - df.loc[index, 'HPPN_Y']) ** 2) ** 0.5
        
        if distance < min_distance:
            min_distance = distance
            nearest_row = row
    
    if nearest_row is None:
        return None
    
    return nearest_row[column_names]

for column in ['일강수량(mm)', '일 최심적설(cm)', '안개 계속시간(hr)']:
    for i, row in f_df_2020[f_df_2020[column].isna()].iterrows():
        nearest_row = get_nearest_row(f_df_2020, i, [column])
        if nearest_row is not None:
            f_df_2020.loc[i, column] = nearest_row[column]
```
